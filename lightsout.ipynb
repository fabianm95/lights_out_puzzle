```python
import numpy as np
import scipy.linalg as ln

np.set_printoptions(threshold=np.inf)
a = np.zeros((64,64))

for i in range(0, 8):
    for j in range(0, 8):
        ma = np.zeros((8,8))
        ma[i][j] = 1
        if(i>0):
            ma[i-1][j] = 1
        if(i<7):
            ma[i+1][j] = 1
        if(j>0):
            ma[i][j-1] = 1
        if(j<7):
            ma[i][j+1] = 1
        a[:,8*i+j] = np.concatenate(ma)
        
def savemat(mat,name):
    s = '{'
    n = mat.shape[0]
    for i in range(0,n):
        s += '{'
        for j in range(0,n):
            s += '%u,' %(mat[i][j])
        s+= '},'
    s = s[:-1]
    s += '}'
    f = open(name,'w')
    f.write(s)

savemat(a,'moves.dat')
    
def gauss(A,b):
    ma = A
    be = b
    n = b.shape[0]
    inv = np.eye(n)
    for j in range(0,n):
        idx = int(np.argwhere(ma[j,:]==1)[0])
        #print type(idx)
        ind = np.argwhere(ma[:,idx]==1)
        #print 'id',ind
        ind = ind[np.where(ind != idx)]
        #print 'idx', idx
        #print 'ind', ind
        for k in ind:
            ma[k,:] = np.int_(np.logical_xor(ma[k,:],ma[idx,:]))
            be[k] = np.int_(np.logical_xor(b[k],b[idx]))
            inv[k,:] = np.int_(np.logical_xor(inv[k,:],inv[idx,:]))
    for j in range(0,n):
        idx = int(np.argwhere(ma[j,:]==1)[0])
        #print type(idx)
        ind = np.argwhere(ma[:,idx]==1)
        #print 'id',ind
        ind = ind[np.where(ind != idx)]
        #print 'idx', idx
        #print 'ind', ind
        for k in ind:
            ma[k,:] = np.int_(np.logical_xor(ma[k,:],ma[idx,:]))
            be[k] = np.int_(np.logical_xor(b[k],b[idx]))
            inv[k,:] = np.int_(np.logical_xor(inv[k,:],inv[idx,:]))
    for j in range(0,n):
        idx = int(np.argwhere(ma[j,:]==1)[0])
        #print type(idx)
        ind = np.argwhere(ma[:,idx]==1)
        #print 'id',ind
        ind = ind[np.where(ind != idx)]
        #print 'idx', idx
        #print 'ind', ind
        for k in ind:
            ma[k,:] = np.int_(np.logical_xor(ma[k,:],ma[idx,:]))
            be[k] = np.int_(np.logical_xor(b[k],b[idx]))
            inv[k,:] = np.int_(np.logical_xor(inv[k,:],inv[idx,:]))
    for j in range(0,n):
        idx = int(np.argwhere(ma[j,:]==1)[0])
        #print type(idx)
        ind = np.argwhere(ma[:,idx]==1)
        #print 'id',ind
        ind = ind[np.where(ind != idx)]
        #print 'idx', idx
        #print 'ind', ind
        for k in ind:
            ma[k,:] = np.int_(np.logical_xor(ma[k,:],ma[idx,:]))
            be[k] = np.int_(np.logical_xor(b[k],b[idx]))
            inv[k,:] = np.int_(np.logical_xor(inv[k,:],inv[idx,:]))
    
    #ma = ma[::-1]
    #be = be[::-1]
    test = ma==np.eye(64)
    return ma, be, test, inv

var, y, p, inv = gauss(a,np.ones(64))
savemat(inv,'inverse.dat')
print inv
```
